## NLP_LAB_7_1_PSG_CFG_GRAMMAR
#### 1. Phrase Structure Grammar (PSG) / Constituency Grammar using NLTK

```python
import nltk
from nltk import CFG
```

```python
# EXAMPLE 1 : Senetence with only Verb phrase as a verb say "Sing"
grammar_ex1_new1 = CFG.fromstring("""
S -> VP
VP -> V
V -> 'Sing'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex1_new1)
# Sentence to parse
sentence = "Sing".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
# Observe that VP is V
```

```python
# EXAMPLE 2 :Sentence with only Noun phrase as "Birla"
grammar_ex1_new2 = CFG.fromstring("""
S -> NP
NP -> N
N -> 'Birla'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex1_new2)
# Sentence to parse
sentence = "Birla".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
# Observe that NP is N
```

```python
# EXAMPLE 3 : Sentence only with Noun phrase = Noun + Adjective say "The melodius song"
grammar_ex1_new3 = CFG.fromstring("""
S -> NP
NP -> Det Adj N | Det N
Det -> 'The'
Adj -> 'melodius'
N -> 'song'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex1_new3)
# Sentence to parse
sentence = "The melodius song".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex2_new1)
# Sentence to parse
sentence = "Eat fast".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 5: Senetnce with NP and VP as 'The cat sleeps'
# Creat the phrase structure (parse tree)
grammar_ex5 = CFG.fromstring("""
  S -> NP VP
  NP -> Det N
  VP -> V
  Det -> 'The'
  N -> 'cat'
  V -> 'sleeps'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex5)
# Parse a sentence
sentence = "The cat sleeps".split()
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 6: Senetnce with two NP and VP with the Determiner 'Aman chases the cat'
grammar_ex6 = CFG.fromstring("""
    S -> NP VP
    NP -> N | Det N
    VP -> V NP|V
    Det -> 'the'
    N -> 'Aman' | 'cat'
    V -> 'chases'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex6)
# Parse a sentence
sentence = "Aman chases the cat".split()
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 9: Sentence with continuous present VP and Pronoun: 'Priti is reading a book'
grammar_ex_9 = CFG.fromstring("""
S -> NP VP
NP -> N | Det N
  # NP can be a pronoun, a determiner + noun, or a proper noun
VP -> Aux V NP | Aux V
Det -> 'a'
N -> 'Priti'|'book'
Aux -> 'is'
V -> 'reading'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex_9)
# Parse a sentence
sentence = "Priti is reading a book".split()
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 10: 'She is reading a book'
grammar_ex10 = CFG.fromstring("""
S -> NP VP
NP -> Pronoun | Det N
VP -> Aux V NP | Aux V
Pronoun -> 'She'
Det -> 'a'
N -> 'book'
Aux -> 'is'
V -> 'reading'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex10)
# Parse a sentence
sentence = "She is reading a book".split()
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 11:Senetnce with a NP & adjective, VP & Adverb:'A beautiful bird sings melodiously'
grammar_ex11 = CFG.fromstring("""
S -> NP VP
NP -> Det Adj N | Det N
VP -> V Adv | V
Det -> 'A'
Adj -> 'beautiful'
N -> 'bird'
V -> 'sings'
Adv -> 'melodiously'
""")
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex11)
# Sentence to parse
sentence = "A beautiful bird sings melodiously".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# Example 12 : Sentence with NP and a conjunction: 'The brown fox is quick and sharp'
grammar_ex12 = CFG.fromstring("""
    S ->  NP VP
    NP -> Det Adj N | Det N | N
    VP -> Aux Adj | Aux Adj Conj Adj
    Det -> 'The'
    Adj -> 'brown' | 'quick' | 'sharp'
    N -> 'fox'
    Conj -> 'and'
    Aux -> 'is'
    """ )
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex12)
# Sentence to parse
sentence = "The brown fox is quick and sharp".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# Example 13 : Small sentence including conjunction between two NPs:
# 'Rajeshwari and Parul are good friends'
grammar_ex13 = CFG.fromstring("""
    S ->  NP VP
    NP -> N Conj N
    VP -> Aux Adj N
    Adj -> 'good'
    N -> 'Rajeshwari' | 'Parul' | 'friends'
    Conj -> 'and'
    Aux -> 'are'
    """ )
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex13)
# Sentence to parse
sentence = "Rajeshwari and Parul are good friends".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 14 :Big Sentence with NP, VP, Det,Adjective, AUX, Pronoun, and a Conjunction
# "The brown fox is quick and he is jumping over the lazy dog"
grammar_ex14 = CFG.fromstring("""
    S -> NP VP Conj S | NP VP
    NP -> Det Adj N | Pronoun | Det Adj N | N
    VP -> Aux Adj | Aux V PP | Aux V NP | V NP PP | V
    PP -> P NP
    Det -> 'The' | 'the'
    Adj -> 'brown' | 'quick' | 'lazy'
    N -> 'fox' | 'dog'
    Conj -> 'and'
    Pronoun -> 'he'
    Aux -> 'is'
    V -> 'jumping'
    P -> 'over'
    """ )
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex14)
# Print Grammar Rules
print("Grammar Rules:\n", parser.grammar())
# Sentence to parse
sentence = "The brown fox is quick and he is jumping over the lazy dog".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```

```python
# EXAMPLE 15 :Big Sentence with NP, VP, Det,Adjective, AUX, Pronoun, and a Conjunction
# "The brown fox is quick and he is jumping over the lazy dog"
# Alternatively we break it into two sentences and then produce PSG for individual sentence
grammar_ex15 = CFG.fromstring("""
    S ->  S Conj S | NP VP
    NP -> Det Adj N | Pronoun | Det N | N
    VP -> Aux Adj | Aux V PP | Aux V NP | V NP PP | V
    PP -> P NP
    Det -> 'The' | 'the'
    Adj -> 'brown' | 'quick' | 'lazy'
    N -> 'fox' | 'dog'
    Conj -> 'and'
    Pronoun -> 'he'
    Aux -> 'is'
    V -> 'jumping'
    P -> 'over'
    """ )
```

```python
# Create a parser
parser = nltk.ChartParser(grammar_ex15)
# Print Grammar Rules
print("Grammar Rules:\n", parser.grammar())
# Sentence to parse
sentence = "The brown fox is quick and he is jumping over the lazy dog".split()
# Parse the sentence and print the PSG tree
for tree in parser.parse(sentence):
  tree.pretty_print()
```
