### 2. Dependency Grammar using spaCy

```python
import nltk
from nltk import CFG
from spacy import displacy
```

```python
import spacy
# Create an object to load the en_core_web_sm vocubulaory
nlp = spacy.load("en_core_web_sm")
```

```python
# EXAMPLE 1
# Sentence parsing
sentence_1 = "Dance"
sentence_2 ='Beautiful'
# Get the tokenized form of the sentence_1, 2
doc_1 = nlp(sentence_1)
doc_2 = nlp(sentence_2)
# Print dependency parsing
for token in doc_1:
    print(f"{token.text} -> {token.dep_} -> {token.head.text}")
for token in doc_2:
    print(f"{token.text} -> {token.dep_} -> {token.head.text}")
# Visualize the dependency graph using displacy
displacy.render(doc_1, style="dep", jupyter=True)
displacy.render(doc_2, style="dep", jupyter=True)
```
