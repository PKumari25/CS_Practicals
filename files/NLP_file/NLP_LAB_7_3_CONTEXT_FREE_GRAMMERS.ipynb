#### 3. Context-Free Grammar (CFG) using NLTK

```python
import nltk
from nltk import CFG
```

```python
# Example 1: Define the CFG grammar for sentence: Amol likes icecream
CFG_1 = CFG.fromstring("""
    S -> NP VP
    NP -> N
    N -> 'Amol' | 'icecream'
    VP -> V NP
    V -> 'likes'
""")
# Create the parser
parser = nltk.ChartParser(CFG_1)
# Print Grammar Rules
print("Grammar Rules:\n", parser.grammar())
# Sentence to parse
sentence = "Amol likes icecream".split()
# Parse and display the tree
for tree in parser.parse(sentence):
    tree.pretty_print()
```

```python
# Example 2: Define the CFG grammar with sentence: Sunidhi loves mangos and apples
CFG_2 = CFG.fromstring("""
    S -> NP VP
    NP -> N|N Conj N
    N -> 'Sunidhi' | 'mangos'|'apples'
    VP -> V NP
    V -> 'loves'
    Conj -> 'and'
""")
# Create the parser
parser = nltk.ChartParser(CFG_2)
# Print Grammar Rules
print("Grammar Rules:\n", parser.grammar())
# Sentence to parse
sentence = "Sunidhi loves mangos and apples".split()
# Parse and display the tree
for tree in parser.parse(sentence):
    tree.pretty_print()
```

```python
# Example 3: Senetnce: The brown fox is quick
CFG_3 = CFG.fromstring("""
    S -> NP VP
    NP -> Det Adj N
    VP -> V Adj
    Det -> 'The'
    Adj -> 'brown' | 'quick'
    N -> 'fox'
    V -> 'is'
""")
# Create parser
parser = nltk.ChartParser(CFG_3)
# Parse sentence
sentence = "The brown fox is quick ".split()
for tree in parser.parse(sentence):
    tree.pretty_print()
```

```python
# Example 4: He is jumping over the lazy dog
CFG_4 = CFG.fromstring("""
    S -> NP VP
    NP -> Pronoun | Det Adj N
    VP -> Aux V PP
    PP -> P NP
    Pronoun -> 'He'
    Aux -> 'is'
    V -> 'jumping'
    P -> 'over'
    Det -> 'the'
    Adj -> 'lazy'
    N -> 'dog'
""")

# Create parser
parser = nltk.ChartParser(CFG_4)

# Sentence to parse
sentence = "He is jumping over the lazy dog".split()

# Parse and display tree
for tree in parser.parse(sentence):
    tree.pretty_print()
```

```python
# Example 5: Define the CFG with senetnce: The brown fox is quick and he is jumping over the lazy dog
CFG_5 = CFG.fromstring("""
    S -> S Conj S
    S -> NP VP
    NP -> Det Adj N | Pronoun | Det Adj Adj N
    VP -> Aux Adj | Aux V PP
    PP -> P NP

    Conj -> 'and'
    Det -> 'The' | 'the'
    Adj -> 'brown' | 'quick' | 'lazy'
    N -> 'fox' | 'dog'
    Pronoun -> 'he'
    Aux -> 'is'
    V -> 'jumping'
    P -> 'over'
""")

# Create the parser
parser = nltk.ChartParser(CFG_5)

# Sentence to parse
sentence = "The brown fox is quick and he is jumping over the lazy dog".split()

# Parse and display the tree
for tree in parser.parse(sentence):
    tree.pretty_print()
```

##### Exclamatory Sentences
##### "What a beautiful picture!!" This sentence is an exclamatory sentence and can be interpreted structurally as:            Sentence Structure Breakdown
##### Interjection/Exclamation (INTJ): What
##### Noun Phrase (NP): a beautiful picture
##### So the sentence can be simplified to:
##### S → INTJ NP

```python
# Example 6: Exclamatory Sentencs: What a beautiful pricture
CFG_6=CFG.fromstring("""
S -> INTJ NP
    NP -> Det Adj N
    INTJ -> 'What'
    Det -> 'a'
    Adj -> 'beautiful'
    N -> 'picture'
""")

# Create the parser
parser = nltk.ChartParser(CFG_6)

# Sentence to parse (strip punctuation for grammar matching)
sentence = "What a beautiful picture".split()

# Parse the sentence
for tree in parser.parse(sentence):
    tree.pretty_print()
```

##### General Interrogative Sentence Structures i.e. a question WH Q
##### Examples of Questions:
##### "What is this?", "Who are you?", "Where is my book?"
##### They follow common interrogative structures:
##### WH-word + Auxiliary Verb + Subject (+ Predicate)
##### We define CFG rules as follows:
##### Q  → WH_AUX NP
##### Q  → WH_AUX NP VP
##### NP → Pronoun | Det N | ProperNoun
##### VP → V NP | V
##### WH_AUX → WH AUX

```python
# Define CFG for basic WH questions
CFG_7 = CFG.fromstring("""
    Q -> WH_AUX NP
    WH_AUX -> WH AUX
    NP -> Pronoun | Det N | ProperNoun
    WH -> 'What' | 'Who' | 'Where'
    AUX -> 'is' | 'are'
    Det -> 'my'
    N -> 'book'
    Pronoun -> 'you' | 'this'
""")

# Create the parser
parser = nltk.ChartParser(CFG_7)

# Sentences to parse
sentences = [
    "What is this",
    "Who are you",
    "Where is my book"
]

# Parse each sentence
for sent in sentences:
    print(f"\nParsing: '{sent}'")
    for tree in parser.parse(sent.split()):
        tree.pretty_print()
```
