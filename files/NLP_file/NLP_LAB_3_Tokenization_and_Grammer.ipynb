### to repeat these calculations on several texts, we define our own name for a task, like “lexical_diversity” or “percentage”, and associate it with a block of code.
### Once we do that, we only have to type a short name instead of one or more complete lines of Python code, and we can reuse it as often as we want. The block of code that does a task for us is called a function.
### Define a short name for the function with the keyword def followed by two new functions, lexical_diversity() and percentage():

```python
# Define a function for finding the lexical diversity
def lexical_diversity(text):
  return len(text) / len(set(text))
```

```python
# Define a function for finding the percentage of the text where the occurance exists
def percentage(count, total):
  return (count / total)*100
```

### Question: Find the Lexical diversity of various genres(tokens, Types, counts, lexical diveristy, percentage) in the Brown Corpus which
### (i) Loads the Brown Corpus from NLTK.
### (ii) Extracts words from each genre.
### (iii) Calculates Tokens: Total words in the genre, Types: Unique words, Type-Token Ratio (TTR): Unique words / Total words, Lexical Diversity Percentage: (Types / Tokens) * 100 and Stores results in a Pandas DataFrame for better readability

```python
from nltk.corpus import brown
import pandas as pd
```

```python
# Ensure Brown Corpus is downloaded
nltk.download('brown')
```

```python
# Function to compute lexical diversity measures
def lexical_diversity(genre):
  words = brown.words(categories=genre)
  tokens = len(words)  # Total words
  types = len(set(words))  # Unique words
  ttr = types / tokens  # Type-token ratio
  lexical_diversity = types / tokens * 100  # Percentage representation
  return tokens, types, ttr, lexical_diversity
```

```python
# Get genres from Brown corpus
genres = brown.categories()
print(genres)
```
